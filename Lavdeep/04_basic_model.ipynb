```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import math

# -------------------------
# LOAD DATA
# -------------------------
X_train = np.load("../data/processed/X_train.npy")
y_train = np.load("../data/processed/y_train.npy")

X_val = np.load("../data/processed/X_val.npy")
y_val = np.load("../data/processed/y_val.npy")

# -------------------------
# FLATTEN SEQUENCES
# (30,16) → (480)
# -------------------------
X_train_flat = X_train.reshape(X_train.shape[0], -1)
X_val_flat = X_val.reshape(X_val.shape[0], -1)

print("Flattened shape:", X_train_flat.shape)

# -------------------------
# TRAIN MODEL
# -------------------------
model = LinearRegression()
model.fit(X_train_flat, y_train)

# -------------------------
# PREDICTIONS
# -------------------------
y_train_pred = model.predict(X_train_flat)
y_val_pred = model.predict(X_val_flat)

# -------------------------
# EVALUATION (RMSE)
# -------------------------
train_rmse = math.sqrt(mean_squared_error(y_train, y_train_pred))
val_rmse = math.sqrt(mean_squared_error(y_val, y_val_pred))

print("\n=== BASELINE RESULTS ===")
print("Train RMSE:", train_rmse)
print("Val RMSE:", val_rmse)

# -------------------------
# SAMPLE PREDICTIONS
# -------------------------
print("\n=== SAMPLE PREDICTIONS ===")
for i in range(5):
    print(f"Actual: {y_val[i]} | Predicted: {round(y_val_pred[i],2)}")
```

```python
X_test = np.load("../data/processed/X_test_final.npy")
rul_test = np.load("../data/processed/rul_test.npy")

X_test_flat = X_test.reshape(X_test.shape[0], -1)

y_test_pred = model.predict(X_test_flat)

test_rmse = math.sqrt(mean_squared_error(rul_test, y_test_pred))
print("Test RMSE:", test_rmse)
```
